 possvel calcular a presso de vapor do ar mido de entrada, sabendo que, a $28^\circ\mathrm{C}$ a presso de saturao da gua vale $3782\,\mathrm{Pa}$. Assim, utiliza-se:

$$
\Phi_e = \frac{p_{v,e}}{p_{sat,e}} \hspace{5mm} \therefore \hspace{5mm} p_{v,e} = \Phi_e p_{sat,e} = 2647,4 \mathrm{Pa}
$$

### Questo 1 b)

Partindo de:

$$
pV = nRT
$$

Substitui-se  $n$ por $m/\bar{M}$ para obter:

$$
pV = \frac{m}{\bar{M}} R T
$$

Assim, calcula-se:

$$
\frac{m}{V} = \frac{p\bar{M}}{RT}
$$

Para a gua dissolvida no ar de entrada, tem-se:

 - $p = 2647,4\,\mathrm{Pa}$
 - $\bar{M} = 18,00\,\mathrm{g/m^3}$
 - $R = 8,314 \,\mathrm{J/mol.K}$
 - $T = 301\,\mathrm{K}$

Assim, a quantidade de gua dissolvida no ar de entrada  de:

$$
\rho_{v,e} = \frac{m}{V} = 19,04 \,\mathrm{g/m^3}
$$

```python
M_v = 18  # g/mol
R = 8.314  # J/mol.K
T = 301    # K

p = p_v_e

rho_v_e = (p*M_v/(R*T))/1000  # kg/m

print(f'Quantidade de gua dissolvida no ar de entrada: {1000*rho_v_e:.2f} g/m')
```

### Questo 1 e)

Para calcular a entalpia volumtrica do ar mido de entrada,  preciso calcular a entalpia relacionada ao ar seco ($H_s$) e ao vapor d'gua dissolvido nele ($H_v$). Das tabelas termodinmicas, para $28^\circ\mathrm{C}$ ou $301\,\mathrm{K}$. Das tabelas, tem-se:

- $h_{s,e} = 301,2 \,\mathrm{kJ/kg}$
- $h_{v,e} = 2552,6 \,\mathrm{kJ/kg}$

A entalpia volumtrica do ar mido de entrada  dada por:

$$
\frac{H_{u,e}}{V} = \rho_{s,e} h_{s,e} + \rho_{v,e} h_{v,e} = 388,05\,\mathrm{kJ/kg}
$$

```python
def interpola(p1, p2, x):
  a = (p2[1] - p1[1])/(p2[0] - p1[0])
  b = p1[1] - a*p1[0]
  return a*x + b

p1 = (300, 300.19)  # K, kJ/kg
p2 = (305, 305.22)  # K, kJ/kg

h_s_e = interpola(p1, p2, 301)

print(f'Entalpia especfica do ar seco: {h_s_e:.2f} kJ/kg')

h_v_e = 2552.6  # kJ/kg

H_u_e_V = rho_s_e*h_s_e + rho_v_e*h_v_e

print(f'Entalpia volumtrica do ar mido de entrada: {H_u_e_V:.2f} kJ/m')
```

```python
h_u_e = H_u_e_V*v_u_e

print(f'Entalpia especfica do ar mido de entrada: {h_u_e:.2f} kJ/kg')
```

### Questo 1 m)

O ar de sada possui $8,7\,\mathrm{g/m^3}$ de gua e $1159,22\,\mathrm{g/m^3}$ de ar seco na mistura. Isso permite calcular a razo de massa de gua por ar seco de $w = 7,51 \,\mathrm{g/kg(ar\, seco)}$.

Alm disso, para esta condio, a densidade de ar seco pode ser calculada na forma:

$$
\frac{m_{s,s}}{V} = \frac{p_{s,s}\bar{M}_s}{RT}
$$

Bem como a densidade de vapor d'gua na mistura pode ser calculada como:

$$
\frac{m_{v,s}}{V} = \frac{p_{v,s}\bar{M}_{v}}{RT}
$$

Ao dividir a segunda expresso pela primeira, se obtm:

$$
w = \frac{m_{v,s}}{m_{s,s}} = \frac{p_{v,s} \bar{M}_v}{p_{s,s}\bar{M}_s}
$$

A razo entre a massa de ar seco e a de gua nesta mistura  constante e independe da temperatura. Assim, tambm tem que ser vlida a relao:

$$
w = \frac{m_{v,i}}{m_{s,i}} = \frac{p_{v,i} \bar{M}_v}{p_{s,i}\bar{M}_s}
$$


Por isso, pode-se dizer que a relao entre as presses do ar e de vapor d'gua  dada por:

$$
p_{s,i} = zp_{v,i}
$$

Sendo o valor de $z$:

$$
z = \frac{m_{s,i}}{m_{v,i}} \frac{\bar{M}_v}{\bar{M}_s} = \frac{\bar{M}_v}{w\bar{M}_s}
$$

Alm disso, como o processo ocorre  presso constante $p = 1\,\mathrm{bar}$, esta deve ser a soma das presses parciais dos gases:

$$
p_{s,i} + p_{v,i} = p
$$

Resolvendo o sistema se obtm:

$$
p_{v,i} = \frac{p}{1+z} = 1194\,\mathrm{Pa}
$$

Porm, $p_{v}$ deve ser a presso de saturao da gua, pois a mistura sai saturada da bobina de refrigerao. Ou seja, $p_{v,i} = p_{sat,i}$.

```python
w = rho_v_s/rho_s_s

print(f'Razo de massa de gua por ar seco: {1000*w:.2f} g/kg(ar seco)')

z = M_v/((w)*M_s)

p_v_i = 1e5/(1+z)

print(f'Presso de saturao da gua: {p_v_i:.2f} Pa')

Ti = interpola((1072, 8), (1228, 10), p_v_i)

print(f'Temperatura de sada da unidade de refrigerao: {Ti:.2f} C')
```

### Questo 1 q)

A massa de vapor d'gua por massa de ar seco na entrada  dada por:

$$
w_e = \frac{\rho_{v,e}}{\rho_{s,e}} = 16,9\,\mathrm{g/kg\,(ar\,seco)}
$$

J na sada da bobina de refrigerao  dada por:

$$
w_i = \frac{\rho_{v,i}}{\rho_{s,i}} = 7,49\,\mathrm{g/kg\,(ar\,seco)}
$$


Ou seja, neste processo se condensa $9,39\,\mathrm{g/kg\,(ar\,seco)}$ de gua. Portanto, neste processo, a vazo mssica de ar mido de entrada pode ser expressa na forma:

$$
\dot{m}_{u,e} = \dot{m}_{s,e}\left(\frac{\dot{m}_{s,e} + \dot{m}_{v,e}}{\dot{m}_{s,e}}\right) = \dot{m}_{s,e}\left(1+w_e\right)
$$


J na sada da bobina de refrigerao, a vazo mssica de ar mido pode ser expressa por:

$$
\dot{m}_{u,i} = \dot{m}_{s,i}\left(\frac{\dot{m}_{s,i} + \dot{m}_{v,i}}{\dot{m}_{s,i}}\right) = \dot{m}_{s,i}\left(1+w_i\right)
$$

E a vazo mssica de vapor d'gua consensado pode ser expressa por:

$$
\dot{m}_{c,i} = \dot{m}_{s,e} (w_e-w_i)
$$

E a entalpia desta gua condensada  dada pela entalpia de lquido saturado da tabela termodinmica a $9,56^\circ\mathrm{C}$:

$$
h_{c,i} = 40,18\,\mathrm{kJ/kg}
$$

Assim, a quantidade de calor trocado ser de:

$$
\dot{q} = \dot{m}_{u,e}h_{u,e} - \dot{m}_{u,i}h_{u,i} - \dot{m}_{c,i}h_{c,i}  
$$

Substituindo os valores encontrados anteriormente:
$$
\dot{q} = \dot{m}_{s,e}\left(1+w_e\right)h_{u,e} - \dot{m}_{s,i}\left(1+w_i\right)h_{u,i} - \dot{m}_{s,e} (w_e-w_i)h_{c,i}  
$$


A massa de ar seco que entra e sai de cada processo  a mesma. Por isso:

$$
\dot{m}_{s,e} = \dot{m}_{s,i} = \dot{m}_{s,s} = \dot{m}_{s}
$$

Assim a expresso anterior se simplifica para:

$$
\dot{q} = \dot{m}_{s}\left[\left(1+w_e\right)h_{u,e} - \left(1+w_i\right)h_{u,i} - (w_e-w_i)h_{c,i}\right]  
$$

Isso permite o clculo de $\dot{m}_s$:

$$
\dot{m}_s = \frac{\dot{q}}{\left(1+w_e\right)h_{u,e} - \left(1+w_i\right)h_{u,i} - (w_e-w_i)h_{c,i}} = 0,9132\,\mathrm{kg/s}
$$

A vazo mssica do ar mido de entrada  dada por:

$$
\dot{m}_{u,e} = \dot{m}_s\left(1+w_e\right) = 0,9287\,\mathrm{kg/s}
$$

```python
w_e = rho_v_e/rho_s_e

print(f'Massa de vapor d\'gua por massa de ar seco na entrada: {1000*w_e:.2f} g/kg (ar seco)')

rho_v_i = 1/v_v_i
rho_s_i = 1/v_s_i

w_i = rho_v_i/rho_s_i

print(f'Massa de vapor d\'gua por massa de ar seco na sada: {1000*w_i:.2f} g/kg (ar seco)')

h_c_i = interpola((8, 33.6), (10, 42.01), Ti)

print(f'Entalpia de lquido saturado na sada da bobina de refrigerao: {h_c_i:.2f} kJ/kg')

dq = 38.7  # kW

dm_s = dq/((1+w_e)*h_u_e - (1+w_i)*h_u_i - (w_e-w_i)*h_c_i)

print(f'Vazo mssica de ar seco: {dm_s:.4f} kg/s')

dm_u_e = dm_s*(1+w_e)

print(f'Vazo mssica de ar mido de entrada: {dm_u_e:.4f} kg/s')
```
