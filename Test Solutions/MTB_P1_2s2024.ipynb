### Questão 1 a)

É possível calcular a pressão de vapor do ar úmido de entrada, sabendo que, a $28^\circ\mathrm{C}$ a pressão de saturação da água vale $3782\,\mathrm{Pa}$. Assim, utiliza-se:

$$
\Phi_e = \frac{p_{v,e}}{p_{sat,e}} \hspace{5mm} \therefore \hspace{5mm} p_{v,e} = \Phi_e p_{sat,e} = 2647,4 \mathrm{Pa}
$$



```python
p_sat_e = 3782  # Pa
phi_e = 0.7     # ---

p_v_e = p_sat_e*phi_e

print(f'p_ve = {p_v_e:.2f} Pa')
```

### Questão 1 b)

Partindo de:

$$
pV = nRT
$$

Substitui-se  $n$ por $m/\bar{M}$ para obter:

$$
pV = \frac{m}{\bar{M}} R T
$$

Assim, calcula-se:

$$
\frac{m}{V} = \frac{p\bar{M}}{RT}
$$

Para a água dissolvida no ar de entrada, tem-se:

 - $p = 2647,4\,\mathrm{Pa}$
 - $\bar{M} = 18,00\,\mathrm{g/m^3}$
 - $R = 8,314 \,\mathrm{J/mol.K}$
 - $T = 301\,\mathrm{K}$

Assim, a quantidade de água dissolvida no ar de entrada é de:

$$
\rho_{v,e} = \frac{m}{V} = 19,04 \,\mathrm{g/m^3}
$$

```python
M_v = 18  # g/mol
R = 8.314  # J/mol.K
T = 301    # K

p = p_v_e

rho_v_e = (p*M_v/(R*T))/1000  # kg/m³

print(f'Quantidade de água dissolvida no ar de entrada: {1000*rho_v_e:.2f} g/m³')
```

### Questão 1 e)

Para calcular a entalpia volumétrica do ar úmido de entrada, é preciso calcular a entalpia relacionada ao ar seco ($H_s$) e ao vapor d'água dissolvido nele ($H_v$). Das tabelas termodinâmicas, para $28^\circ\mathrm{C}$ ou $301\,\mathrm{K}$. Das tabelas, tem-se:

- $h_{s,e} = 301,2 \,\mathrm{kJ/kg}$
- $h_{v,e} = 2552,6 \,\mathrm{kJ/kg}$

A entalpia volumétrica do ar úmido de entrada é dada por:

$$
\frac{H_{u,e}}{V} = \rho_{s,e} h_{s,e} + \rho_{v,e} h_{v,e} = 388,05\,\mathrm{kJ/kg}
$$

```python
def interpola(p1, p2, x):
  a = (p2[1] - p1[1])/(p2[0] - p1[0])
  b = p1[1] - a*p1[0]
  return a*x + b

p1 = (300, 300.19)  # K, kJ/kg
p2 = (305, 305.22)  # K, kJ/kg

h_s_e = interpola(p1, p2, 301)

print(f'Entalpia específica do ar seco: {h_s_e:.2f} kJ/kg')

h_v_e = 2552.6  # kJ/kg

H_u_e_V = rho_s_e*h_s_e + rho_v_e*h_v_e

print(f'Entalpia volumétrica do ar úmido de entrada: {H_u_e_V:.2f} kJ/m³')
```

### Questão 1 m)

O ar de saída possui $8,7\,\mathrm{g/m^3}$ de água e $1159,22\,\mathrm{g/m^3}$ de ar seco na mistura. Isso permite calcular a razão de massa de água por ar seco de $w = 7,51 \,\mathrm{g/kg(ar\, seco)}$.

Além disso, para esta condição, a densidade de ar seco pode ser calculada na forma:

$$
\frac{m_{s,s}}{V} = \frac{p_{s,s}\bar{M}_s}{RT}
$$

Bem como a densidade de vapor d'água na mistura pode ser calculada como:

$$
\frac{m_{v,s}}{V} = \frac{p_{v,s}\bar{M}_{v}}{RT}
$$

Ao dividir a segunda expressão pela primeira, se obtém:

$$
w = \frac{m_{v,s}}{m_{s,s}} = \frac{p_{v,s} \bar{M}_v}{p_{s,s}\bar{M}_s}
$$

A razão entre a massa de ar seco e a de água nesta mistura é constante e independe da temperatura. Assim, também tem que ser válida a relação:

$$
w = \frac{m_{v,i}}{m_{s,i}} = \frac{p_{v,i} \bar{M}_v}{p_{s,i}\bar{M}_s}
$$


Por isso, pode-se dizer que a relação entre as pressões do ar e de vapor d'água é dada por:

$$
p_{s,i} = zp_{v,i}
$$

Sendo o valor de $z$:

$$
z = \frac{m_{s,i}}{m_{v,i}} \frac{\bar{M}_v}{\bar{M}_s} = \frac{\bar{M}_v}{w\bar{M}_s}
$$

Além disso, como o processo ocorre à pressão constante $p = 1\,\mathrm{bar}$, esta deve ser a soma das pressões parciais dos gases:

$$
p_{s,i} + p_{v,i} = p
$$

Resolvendo o sistema se obtém:

$$
p_{v,i} = \frac{p}{1+z} = 1194\,\mathrm{Pa}
$$

Porém, $p_{v}$ deve ser a pressão de saturação da água, pois a mistura sai saturada da bobina de refrigeração. Ou seja, $p_{v,i} = p_{sat,i}$.

```python
w = rho_v_s/rho_s_s

print(f'Razão de massa de água por ar seco: {1000*w:.2f} g/kg(ar seco)')

z = M_v/((w)*M_s)

p_v_i = 1e5/(1+z)

print(f'Pressão de saturação da água: {p_v_i:.2f} Pa')

Ti = interpola((1072, 8), (1228, 10), p_v_i)

print(f'Temperatura de saída da unidade de refrigeração: {Ti:.2f} °C')
```

### Questão 1 q)

A massa de vapor d'água por massa de ar seco na entrada é dada por:

$$
w_e = \frac{\rho_{v,e}}{\rho_{s,e}} = 16,9\,\mathrm{g/kg\,(ar\,seco)}
$$

Já na saída da bobina de refrigeração é dada por:

$$
w_i = \frac{\rho_{v,i}}{\rho_{s,i}} = 7,49\,\mathrm{g/kg\,(ar\,seco)}
$$


Ou seja, neste processo se condensa $9,39\,\mathrm{g/kg\,(ar\,seco)}$ de água. Portanto, neste processo, a vazão mássica de ar úmido de entrada pode ser expressa na forma:

$$
\dot{m}_{u,e} = \dot{m}_{s,e}\left(\frac{\dot{m}_{s,e} + \dot{m}_{v,e}}{\dot{m}_{s,e}}\right) = \dot{m}_{s,e}\left(1+w_e\right)
$$


Já na saída da bobina de refrigeração, a vazão mássica de ar úmido pode ser expressa por:

$$
\dot{m}_{u,i} = \dot{m}_{s,i}\left(\frac{\dot{m}_{s,i} + \dot{m}_{v,i}}{\dot{m}_{s,i}}\right) = \dot{m}_{s,i}\left(1+w_i\right)
$$

E a vazão mássica de vapor d'água consensado pode ser expressa por:

$$
\dot{m}_{c,i} = \dot{m}_{s,e} (w_e-w_i)
$$

E a entalpia desta água condensada é dada pela entalpia de líquido saturado da tabela termodinâmica a $9,56^\circ\mathrm{C}$:

$$
h_{c,i} = 40,18\,\mathrm{kJ/kg}
$$

Assim, a quantidade de calor trocado será de:

$$
\dot{q} = \dot{m}_{u,e}h_{u,e} - \dot{m}_{u,i}h_{u,i} - \dot{m}_{c,i}h_{c,i}  
$$

Substituindo os valores encontrados anteriormente:
$$
\dot{q} = \dot{m}_{s,e}\left(1+w_e\right)h_{u,e} - \dot{m}_{s,i}\left(1+w_i\right)h_{u,i} - \dot{m}_{s,e} (w_e-w_i)h_{c,i}  
$$


A massa de ar seco que entra e sai de cada processo é a mesma. Por isso:

$$
\dot{m}_{s,e} = \dot{m}_{s,i} = \dot{m}_{s,s} = \dot{m}_{s}
$$

Assim a expressão anterior se simplifica para:

$$
\dot{q} = \dot{m}_{s}\left[\left(1+w_e\right)h_{u,e} - \left(1+w_i\right)h_{u,i} - (w_e-w_i)h_{c,i}\right]  
$$

Isso permite o cálculo de $\dot{m}_s$:

$$
\dot{m}_s = \frac{\dot{q}}{\left(1+w_e\right)h_{u,e} - \left(1+w_i\right)h_{u,i} - (w_e-w_i)h_{c,i}} = 0,9132\,\mathrm{kg/s}
$$

A vazão mássica do ar úmido de entrada é dada por:

$$
\dot{m}_{u,e} = \dot{m}_s\left(1+w_e\right) = 0,9287\,\mathrm{kg/s}
$$

```python
w_e = rho_v_e/rho_s_e

print(f'Massa de vapor d\'água por massa de ar seco na entrada: {1000*w_e:.2f} g/kg (ar seco)')

rho_v_i = 1/v_v_i
rho_s_i = 1/v_s_i

w_i = rho_v_i/rho_s_i

print(f'Massa de vapor d\'água por massa de ar seco na saída: {1000*w_i:.2f} g/kg (ar seco)')

h_c_i = interpola((8, 33.6), (10, 42.01), Ti)

print(f'Entalpia de líquido saturado na saída da bobina de refrigeração: {h_c_i:.2f} kJ/kg')

dq = 38.7  # kW

dm_s = dq/((1+w_e)*h_u_e - (1+w_i)*h_u_i - (w_e-w_i)*h_c_i)

print(f'Vazão mássica de ar seco: {dm_s:.4f} kg/s')

dm_u_e = dm_s*(1+w_e)

print(f'Vazão mássica de ar úmido de entrada: {dm_u_e:.4f} kg/s')
```
